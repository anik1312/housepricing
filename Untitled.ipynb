```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
cal = fetch_california_housing()
```

##preparing dataset

```python
dataset = pd.DataFrame(cal.data,columns = cal.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=cal.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
## summarize the data
dataset.describe()
```

```python
## explanatory data analysis
## correaltion(1st step)
dataset.corr()
```

```python
plt.scatter(dataset['MedInc'],dataset['Price'])
```

```python
import seaborn as sns
sns.regplot(x="MedInc", y="Price", data=dataset)
```

```python
# indepent and dependent
X= dataset.iloc[:,:-1]
Y= dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

#model training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#print the coefficients and intercepts
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
##predict with test data
mod_predict=model.predict(X_test)
mod_predict
```

```python
#scatter plot for prediction
plt.scatter(Y_test,mod_predict)
```

```python
# residual means error
residuals = Y_test-mod_predict
residuals
```

```python
#plot resiuals
sns.displot(residuals,kind="kde")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(Y_test,mod_predict))
print(mean_squared_error(Y_test,mod_predict))
print(np.sqrt(mean_squared_error(Y_test,mod_predict)))
```

## r squared and adjusted r squared 
r^2=1-ssr/sst
r^2= coeff of dtermination ssr = sum of squares of residual sst=total sum of squares 

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,mod_predict)
print(r2)
```

```python
n = len(Y_test)      # Number of observations
p = X_test.shape[1]  # Number of features

adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("Adjusted R²:", adj_r2)
```
